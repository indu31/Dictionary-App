# The aim of this project is to create an English Dictionary app that returns definitions of English words. When the application is started, it displays a menu as follows:

## Main Menu
            1. Add a new word
            2. Find the meaning
            3. Update a word
            4. Exit
            
                Enter Choice:

### If item 1 is selected, it should prompt for a word and then accept its meaning. This pair of word and its meaning should be stored in a file called ‘words.txt’. When item 2 is selected, it should prompt for a word, search that word in the file ‘words.txt’ and return the meaning if it is found. If the meaning is not found, it should display an appropriate message. Similarly, when choice 3 is entered, it should accept a word and update its meaning. In all the above three cases, the prompt should return to the main menu. The words and their meanings should be stored as a dictionary in words.txt, using the concept of serialization. If choice 4 is selected, a graceful exit should be performed.

```python
import json

Dictionary = {'hello':'greeting',
              'encourage':'inspire,motivate',
              'decision':'conclusion',
              'learn':'memorize',
              'python':'snake,computer language'
             }

j = json.dumps(Dictionary , indent = 4)

with open("Words.txt","w") as f:
    f.write(j)
```

```python
import json
from difflib import get_close_matches

Title = "DICTIONARY APPLICATION"

print(Title.center(200),'\n')

print('~'*120)
print('\n')
print("Main Menu")
print('\t1. Add a new word')
print('\t2. Find The meaning')
print('\t3. Update a word')
print('\t4. Exit')
print('\n')
option = int(input('\t\tEnter the choice : '))

print('~'*120)

while option != 0:
    if option == 1:
        print(option,'. Add New Word')
        print('-'*23)
        n = int(input('Enter the number of words you want to add : '))
        for i in range(n):
            word = input('Enter the new word : ')
            meaning = input('Enter the meaning of new word : ')
            with open("Words.txt","r+") as f:
                data = json.loads(f.read())
                data.update({word : meaning})
            
            with open("Words.txt","w") as f:
                json.dump(data, f, indent = 4)
            
        print('Thank you for adding new word and its meaning.',end='\n')
        print('*'*103)
        print('\nMain Menu')
        print('\t1. Add a new word')
        print('\t2. Find The meaning')
        print('\t3. Update a word')
        print('\t4. Exit')
        
        option = int(input('\t\tEnter the choice : '))
        print('\n')
        print('*'*103)
                
    elif option == 2:
        print(option,'. Find the meaning of the word ')
        print('-'*40)
        
        word = input('Enter the word to find meaning : ')
        with open ("Words.txt", "r") as f:
            data = json.loads(f.read())
            
        word = word.lower()
        if word in data:
            print('\nMeaning of word is : ', data[word])
        elif word.title() in data:
            print('\nMeaning of word is : ', data[word.title()])
        elif word.upper() in data:
            print('\nMeaning of word is : ', data[word.upper()])
        elif len(get_close_matches(word, data.keys())) >0:
            similar_words_list = list(map(str, get_close_matches(word, data.keys())))
            # asking users if they mean any other word form the list of similar_words_list
            ans = input("Did you mean %s instead? Enter 'Y' If yes or 'N' if No " % similar_words_list)
            if ans.lower() == 'y':
                # asking user to select the word
                index = input("Enter the position number of word to select the word from above list. Ex 1 or 2 or 3 : ")
                # here index is for the words displayed
                print(get_close_matches(word, data.keys())[int(index)-1])
            elif ans.lower() == 'n':
                print("\nWord Doesnt exists. Please double check it!!!")
            else:
                print("\nSorry, We don't understand you!!!!")
        else:
            print("\nWord Doesnt exists. Please double check it!!!")

        print('*'*103)
        print('\nMain Menu')
        print('\t1. Add a new word')
        print('\t2. Find The meaning')
        print('\t3. Update a word')
        print('\t4. Exit')
        
        option = int(input('\t\tEnter the choice : '))
        print('\n')
        print('*'*103)
            
    elif option == 3:
        print(option,'. Update a word')
        print('-'*22)
        word = input('Enter the word to update : ')
        meaning = input('Enter the new meaning of word : ')
        print('\n')
        with open ("Words.txt", "r") as f:
            data = json.loads(f.read())
        word = word.lower()
        if word == word.lower() or word == word.title() or word == word.upper():
            data[word] = meaning
        else:
            print("Enter Correct Word")
        with open("Words.txt","w") as f:
                json.dump(data, f, indent = 4)  
                
        print('*'*103)
        print('\nMain Menu')
        print('\t1. Add a new word')
        print('\t2. Find The meaning')
        print('\t3. Update a word')
        print('\t4. Exit')
       
        option = int(input('\t\tEnter the choice : '))
        print('\n')
        print('*'*103)
        
    elif option == 4:
        print(option,'. Exit The Application')
        print('-'*30)
        w = open("Words.txt", "r")
        n1 = input('Enter yes or no or (y/n) to exit the application : ')
        if n1 == 'yes'  or n1 == 'y' or n1 == 'Y' or n1 == 'YES' or n1 == 'Yes':
            w.close()
            print('\nYou have successfully exited application')
            break
        elif n1 == 'no' or n1 == 'n' or n1 == 'N' or n1 == 'NO' or n1 == 'No':
            print('\nYou have not exited the application')
            print('\nEnter the option to continue the application')
            print('*'*103)
            print('\nMain Menu')
            print('\t1. Add a new word')
            print('\t2. Find The meaning')
            print('\t3. Update a word')
            print('\t4. Exit')
            option = int(input('\t\tEnter the choice : '))
            print('\n')
            print('*'*103)
            
```
